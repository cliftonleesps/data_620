```python
import networkx as nx
import networkx.generators.small
import matplotlib.pyplot as plt
```

```python
# Just use the Krackhardt generator
g = networkx.generators.small.krackhardt_kite_graph()
```

Create a dictionary to remap the nodes

```python
mapping = {0:"Andre",
           1:"Beverley",
           2:"Carol",
           3:"Diane",
           4:"Ed",
           5:"Fernando",
           6:"Garth",
           7:"Heather",
           8:"Ike",
           9:"Jane"}
```

```python
options = {
    "font_size": 16,
    "font_weight": "bold",
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 1,
}
```

```python
g=nx.relabel_nodes(g, mapping, copy=True)
```

```python
nx.draw_networkx(g, **options)
plt.axis("off")
plt.show()
```
